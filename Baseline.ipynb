```python
from collections import defaultdict
import csv

def readCSV(path):
  f = open(path, 'rt')
  f.readline()
  for l in f:
    yield l.strip().split(',')
```

## Rating Prediction - Baseline

```python
### Rating baseline: compute averages for each user, or return the global average if we've never seen the user before

allRatings = []
userRatings = defaultdict(list)

for user,book,r in readCSV("assignment1/train_Interactions.csv"):
  r = int(r)
  allRatings.append(r)
  userRatings[user].append(r)
```

```python
globalAverage = sum(allRatings) / len(allRatings)
print(globalAverage)

userAverage = {}
for u in userRatings:
    userAverage[u] = sum(userRatings[u]) / len(userRatings[u])
```

```python
i = 0
for u in userRatings:
    i += 1
    if i > 10:
        break
    print (u, userRatings[u])

print(len(userAverage))
```

```python
### Rating baseline: compute averages for each user, or return 
### the global average if we've never seen the user before

predictions = open("assignment1/predictions_Rating.txt", 'w')

for l in open("assignment1/pairs_Rating.txt"):
  # write header  
  if l.startswith("userID"):
    predictions.write(l)
    continue
    
  # write user-item-rating
  u,b = l.strip().split('-')
  if u in userAverage:
    predictions.write(u + '-' + b + ',' + str(userAverage[u]) + '\n')
  else:
    predictions.write(u + '-' + b + ',' + str(globalAverage) + '\n')

predictions.close()
```

## Read Prediction - Baseline

```python
bookCount = defaultdict(int)
totalRead = 0

for user,book,_ in readCSV("assignment1/train_Interactions.csv"):
  bookCount[book] += 1
  totalRead += 1

mostPopular = [(bookCount[x], x) for x in bookCount]
mostPopular.sort()
mostPopular.reverse()
```

```python
return1 = set()
count = 0
for ic, i in mostPopular:
  count += ic
  return1.add(i)
  if count > totalRead/2: break
```

```python
### Would-read baseline: just rank which books are popular and which are not, 
### and return '1' if a book is among the top-ranked

predictions = open("assignment1/predictions_Read.txt", 'w')
for l in open("assignment1/pairs_Read.txt"):
  if l.startswith("userID"):
    #header
    predictions.write(l)
    continue
  u,b = l.strip().split('-')
  if b in return1:
    predictions.write(u + '-' + b + ",1\n")
  else:
    predictions.write(u + '-' + b + ",0\n")

predictions.close()
```

## Catagory Prediction - Baseline

```python
catDict = {
  "children": 0,
  "comics_graphic": 1,
  "fantasy_paranormal": 2,
  "mystery_thriller_crime": 3,
  "young_adult": 4
}
```

```python
def readfile(path):
  for l in open(path, 'rt'):
    yield eval(l)
```

```python
### Category prediction baseline: Just consider some of the most common words from each category

predictions = open("assignment1/predictions_Category.txt", 'w')
predictions.write("assignment1/userID-reviewID,prediction\n")

for l in readfile("assignment1/test_Category.json"):
  cat = catDict['fantasy_paranormal'] # If there's no evidence, just choose the most common category in the dataset
  words = l['review_text'].lower()
  if 'children' in words:
    cat = catDict['children']
  if 'comic' in words:
    cat = catDict['comics_graphic']
  if 'fantasy' in words:
    cat = catDict['fantasy_paranormal']
  if 'mystery' in words:
    cat = catDict['mystery_thriller_crime']
  if 'young' in words:
    cat = catDict['young_adult']
  predictions.write(l['user_id'] + '-' + l['review_id'] + "," + str(cat) + "\n")

predictions.close()
```
